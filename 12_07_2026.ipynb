```python
from matplotlib.lines import lineStyles
# Data Visualization using Matlotlib Library
# 1. How has startup funding trend changed over the year?
import matplotlib.pyplot as plt

demo = data.groupby("year")["amount"].sum()
plt.figure(figsize=(10,6))

plt.plot(demo.index, demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total Amount")
plt.title("Year wise amount investment",color="red",fontsize=16)
plt.grid(linestyle="--",color="skyblue",linewidth=0.5)
plt.xlabel("Year",color = "blue",fontsize=12)
plt.ylabel("Total Amount", color="blue",fontsize=12)
plt.legend()
plt.show()
```

```python
# 2.Which cities have the highest number of startup?
Demo = data["city"].value_counts().head(10)
color = ["blue","black","red","yellow","pink","purple","orange","brown","skyblue","black"]
plt.figure(figsize=(10,6))
plt.bar(Demo.index,Demo.values,color=color,alpha=0.4,edgecolor="black")

plt.xticks(rotation = 45)

plt.title("Top 10 startup company's in India")
plt.xlabel("City Names")
plt.ylabel("Number of start up")

plt.grid(True)
```

```python
from matplotlib.typing import LineStyleType
#3.which startup sectors have the higest number of funded startups?
#3) Which startup sectors have the highest number of funded startup?
demo=data["vertical"].value_counts().head(10)
colors=["#499A13","#BD4444","#00E0BA","#458393","#293681","#443199","#6FCF97","#AD5C71","#AE2448","#2C687B"]

plt.figure(figsize=(10,6))

plt.barh(demo.index[::-1],demo.values[::-1],color=colors,edgecolor="grey")
plt.xticks(rotation = 45)
plt.title("Top 10 funded startup comapny's in India")
plt.xlabel("Startup Names",color="#005461",fontsize=20)
plt.ylabel(" startup",color="#005461",fontsize=12)
plt.grid(axis='x',linestyle="--",color="grey")


plt.show()

```

```python
# 4.What is the distribution of different fuding rounds?
Demo=data["round"].value_counts().head(7)
plt.figure(figsize=(10,6))


color = ["orange","white","green","black","yellow","blue","grey"]
plt.pie(Demo.values,labels = Demo.index,autopct="%1.1f%%",colors=color)
plt.title("Rounds disttribution in percentage")
```

```python
# 5.How are startup funding amounts distributed?
Demo=data["amount"].value_counts().head(7)
plt.hist(data["amount"],bins=10,edgecolor="black")
```

```python
# 6. Distribution od startup funding Year and funding Amount?
plt.figure(figsize=(10,6))
plt.scatter(data["year"],data["amount"],color="yellow")
plt.title("Funding amount by year")
plt.xlabel("year")
plt.ylabel("amount")
plt.grid(linestyle="--")
plt.show()
```

```python
# 7.Which funding round occurs most frequently?
import seaborn as sns
Demo=data["round"].value_counts().head(10).index
plt.figure(figsize=(10,6))
sns.countplot(data, y = "round",order = Demo,palette="viridis")
plt.title("Most frquently founding round")

```
